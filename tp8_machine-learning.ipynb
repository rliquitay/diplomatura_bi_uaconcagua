```
import sklearn 
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

## Codificación de variables no numéricas

```
convertir = LabelEncoder()
```

```
#Convierto todas la columnas con variables no numéricas a string
analisis.mes_ultimo_contactocliente = analisis.mes_ultimo_contactocliente.apply(str)
analisis.rdo_campaña_anterior = analisis.rdo_campaña_anterior.apply(str)
analisis.ultimo_contacto_cliente = analisis.ultimo_contacto_cliente.apply(str)
analisis.es_propietario = analisis.es_propietario.apply(str)
analisis.tiene_deuda = analisis.tiene_deuda.apply(str)
analisis.educacion = analisis.educacion.apply(str)
analisis.estado_civil = analisis.estado_civil.apply(str)
analisis.trabajo = analisis.trabajo.apply(str)
analisis.tiene_prestamo = analisis.tiene_prestamo.apply(str)
analisis.suscribe_deposito = analisis.suscribe_deposito.apply(str)
```

```
#Agrego una columna numérica por cada columna string
analisis.insert(2,column="numtrabajo",value=convertir.fit_transform(analisis.trabajo))
analisis.insert(4,column="numestado_civil",value=convertir.fit_transform(analisis.estado_civil))
analisis.insert(6,column="numeducacion",value=convertir.fit_transform(analisis.educacion))
analisis.insert(8,column="numtiene_deuda",value=convertir.fit_transform(analisis.tiene_deuda))
analisis.insert(11,column="numes_propietario",value=convertir.fit_transform(analisis.es_propietario))
analisis.insert(13,column="numtiene_prestamo",value=convertir.fit_transform(analisis.tiene_prestamo))
analisis.insert(15,column="numcontact",value=convertir.fit_transform(analisis.ultimo_contacto_cliente))
analisis.insert(24,column="numrdo_campañaanterior",value=convertir.fit_transform(analisis.rdo_campaña_anterior))
analisis.insert(25,column="numy",value=convertir.fit_transform(analisis.suscribe_deposito))
```

```
analisis.mes_ultimo_contactocliente.unique()
```

```
def mes(mes_ultimo_contactocliente): #convierto string en número para la variable 'mes del último contacto con el cliente'
  if mes_ultimo_contactocliente == "jan":
    return int("1")
  elif mes_ultimo_contactocliente == "feb":
    return int("2")
  elif mes_ultimo_contactocliente == "mar":
    return int("3")
  elif mes_ultimo_contactocliente == "apr":
    return int("4")
  elif mes_ultimo_contactocliente == "may":
    return int("5")
  elif mes_ultimo_contactocliente == "jun":
    return int("6")
  elif mes_ultimo_contactocliente == "jul":
    return int("7")
  elif mes_ultimo_contactocliente == "aug":
    return int("8")
  elif mes_ultimo_contactocliente == "sep":
    return int("9")
  elif mes_ultimo_contactocliente == "oct":
    return int("10")
  elif mes_ultimo_contactocliente == "nov":
    return int("11")
  else: 
    return int("12")
```

```
analisis.mes_ultimo_contactocliente = analisis.mes_ultimo_contactocliente.apply(mes)
```
